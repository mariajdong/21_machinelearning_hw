```python
# other dependencies
import pandas as pd
import numpy as np
```

## Read the CSV and Perform Basic Data Cleaning

```python
# import data
df = pd.read_csv("exoplanet_data.csv")

# drop null columns & rows
df = df.dropna(axis='columns', how='all')
df = df.dropna()

df.head()
```

## Select your features (columns)

```python
# set features
X = df.drop("koi_disposition", axis=1)
X
```

## Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df['koi_disposition']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, stratify = y)
X_train.head()
```

## Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# scale X data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Train the Model



```python
# create SVC model
from sklearn.svm import SVC 
model = SVC(kernel = 'linear')
model.fit(X_train_scaled, y_train)

print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

## Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# create gridsearch model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}

grid = GridSearchCV(model, param_grid, verbose = 3)
```

```python
# train model w/ gridsearch
grid.fit(X_train_scaled, y_train)
```

```python
# print best params & score
print(grid.best_params_)
print(grid.best_score_)
```

## Save the Model

```python
import joblib

filename = 'model_1.sav'
joblib.dump(model, filename)
```

# random forest
warning for graders: this cell took a *very* long time to run; it had to be left overnight

```python
# create random forest classifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import uniform, truncnorm, randint

rf = RandomForestClassifier(random_state = 42)

rf_params = {'n_estimators': randint(4,200),
             # normally distributed max_features, with mean .25 stddev 0.1, bounded between 0 and 1
             'max_features': truncnorm(a=0, b=1, loc=0.25, scale=0.1),
             # uniform distribution from 0.01 to 0.2 (0.01 + 0.199)
             'min_samples_split': uniform(0.01, 0.199)}

clf = RandomizedSearchCV (rf, rf_params, n_iter=100, cv=5, random_state = 42)

rf_model = clf.fit(X_train_scaled, y_train)
```

```python
rf_model.score(X_test_scaled, y_test)
```

```python
print(rf_model.best_params_)
print(rf_model.best_score_)
```

```python
filename = 'model_2.sav'
joblib.dump(rf_model, filename)
```

# logistic regression

```python
param_grid2 = {'C': [1, 5, 10, 50],
              'penalty': ['l1', 'l2']}

clf2 = GridSearchCV(log, param_grid2, cv = 5, verbose = 3)
```

```python
log_model = clf2.fit(X_train_scaled, y_train)
```

```python
print(log_model.best_params_)
print(log_model.best_score_)
```
